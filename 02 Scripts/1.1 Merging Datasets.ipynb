## 03. Harmonizing and merging data

### 03.01 Harmonizing the 'country' variable

The WDI data contains a lot of geographic entities that are not actually countries, such as 'East Asia & Pacific' or 'Early-demographic dividend'. These should be removed from the dataset before mergin with the other dataframes. To identify the non-country entities I loaded the CSV into Claude and asked it to produce a list of non-country entities. I then adjusted this list manually to keep entities like 'West Bank and Gaza' or 'Taiwan'. 

```python
# List of non-country designations to remove
non_countries = [
    # Regional groupings
    'East Asia & Pacific',
    'East Asia & Pacific (excluding high income)',
    'East Asia & Pacific (IDA & IBRD countries)',
    'Europe & Central Asia',
    'Europe & Central Asia (excluding high income)',
    'Europe & Central Asia (IDA & IBRD countries)',
    'Euro area',
    'European Union',
    'Latin America & Caribbean',
    'Latin America & Caribbean (excluding high income)',
    'Latin America & the Caribbean (IDA & IBRD countries)',
    'Middle East & North Africa',
    'Middle East & North Africa (excluding high income)',
    'Middle East & North Africa (IDA & IBRD countries)',
    'North America',
    'South Asia',
    'South Asia (IDA & IBRD)',
    'Sub-Saharan Africa',
    'Sub-Saharan Africa (excluding high income)',
    'Sub-Saharan Africa (IDA & IBRD countries)',
    'Central Europe and the Baltics',
    'World',
    
    # Development status groups
    'Early-demographic dividend',
    'Post-demographic dividend',
    'Late-demographic dividend',
    'High income',
    'Low & middle income',
    'Low income',
    'Lower middle income',
    'Upper middle income',
    'Heavily indebted poor countries (HIPC)',
    'Least developed countries: UN classification',
    'Small states',
    'Other small states',
    'Pacific island small states',
    
    # Other
    'Macao SAR, China',
    
    # IDA/IBRD groupings
    'IBRD only',
    'IDA & IBRD total',
    'IDA blend',
    'IDA only',
    'IDA total',
    
    # Ensure any spaces or variations are caught
    'IDA & IBRD',
    'IBRD',
    'IDA'
]
```

```python
non_countries
```

```python
#Create 'country_mask' that does NOT contain list 'non_countries'
country_mask = ~wdi_df['Country'].isin(non_countries)
```

```python
country_mask
```

```python
# Apply the mask to get only country entries
wdi_cleaned_df = wdi_df[country_mask].copy()
```

```python
wdi_cleaned_df
```

We reduced the df from 3591 rows to 2937 rows.

Check for additional variations between country names in both dataframes

```python
# Get unique country names from each dataframe
eiu_countries = set(eiu_df['Country'].unique())
wdi_countries = set(wdi_cleaned_df['Country'].unique())
```

```python
# Find countries that are in EIU but not in cleaned WDI
only_in_eiu = eiu_countries - wdi_countries
```

```python
sorted(only_in_eiu)  # sorted() makes the output easier to read
```

```python
# Find countries that are in WDI but not in EIU
only_in_wdi = wdi_countries - eiu_countries
sorted(only_in_wdi)
```

WDI still contains more entities that EIU. But we can see that there are no variations of spellings, which would be the bigger issue. We also know from looking at the source data in Excel that the countrys that are only part of EIU area actually missing from the WDI dataset. A left join should merge all neccessary data at this point.

### 03.02. Merging EIU and WDI data

```python
#Checking dtypes
eiu_df.dtypes
```

```python
wdi_cleaned_df.dtypes
```

```python
# Perform left join on Country and Year for EIU and WDI
merged_eiu_wdi_df = pd.merge(
    left=eiu_df,
    right=wdi_cleaned_df,
    how='left',
    on=['Country', 'Year']
)
```
